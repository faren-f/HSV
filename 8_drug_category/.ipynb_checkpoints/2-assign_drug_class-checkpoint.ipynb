```python
import os
import glob
import json
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import requests
```

```python
workdir = os.path.dirname(os.getcwd())
input_dir = Path(f"{workdir}/8_drug_category/data/")
module_path = Path("/home/bbc8731/HSV/3_module_expansion/data/categories_methods")
drugbank_atc = pd.read_csv(os.path.join(input_dir, "drugbank_atc_level2.csv"))
drugbank_atc
```

```python
module_files = sorted(module_path.rglob("*/drug_repurposing/trustrank/uniprot_ppi/validation/approved_drugs/candidate_drugs_scored.csv"))
for p in module_files:
    base_path = p.parents[0]
    print(base_path)
    test_drugs = pd.read_csv(p)
    test_drugs["drugbank_id"] = test_drugs["drug_id"].str.replace("drugbank.", "", regex=False)
    test_drugs = test_drugs[["drugbank_id", "score", "neg_log_score", "trustrank"]]

    merged_df = test_drugs.merge(drugbank_atc, on="drugbank_id", how="left")
    output_file = base_path / f"candidate_drugs_scored_atc_code.csv"
    merged_df.to_csv(output_file, index=False)
```
