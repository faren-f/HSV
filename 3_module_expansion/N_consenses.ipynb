```python
from pathlib import Path
import pandas as pd
```

```python
path = Path("/home/bbc8731/HSV/3_module_expansion/data/categories_methods/BP_Assembly_and_Packaging/consensus/")
```

```python
dirs = [d.name for d in path.glob("*.tsv")]
gene_count = {}

for i in dirs:
    df = pd.read_csv(path/i, sep="\t")
    genes = df.loc[df['n_methods'] > 3, 'gene'].tolist()
    for g in genes:
        if g in gene_count.keys():
            gene_count[g] += 1
        else:
            gene_count[g] = 1
```
