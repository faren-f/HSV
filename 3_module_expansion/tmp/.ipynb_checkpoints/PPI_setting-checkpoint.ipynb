```python
import os
import pandas as pd

workdir = os.path.dirname(os.getcwd())
nextflow_dir = os.path.join(os.path.dirname(os.path.dirname(os.getcwd())), "diseasemodulediscovery")
og_ppi = pd.read_csv(os.path.join(nextflow_dir, "tests/uniprot_ppi.csv"), header=None)
```

```python
og_ppi.columns = ['source', 'target']
og_ppi
```

```python
uniprot_ids = set(og_ppi[['source', 'target']].values.ravel('K'))
print(len(uniprot_ids))
```

```python
import mygene

# Initialize MyGene
mg = mygene.MyGeneInfo()

# Query MyGene.info to get Ensembl Gene IDs (used by GENCODE)
results = mg.querymany(uniprot_ids, scopes='uniprot', fields='ensembl.gene', species='human')

# Convert results to DataFrame
df_mapping = pd.DataFrame(results)

# Extract Ensembl Gene IDs
def extract_ensembl_gene_id(entry):
    if isinstance(entry, list):
        return entry[0]['gene']
    elif isinstance(entry, dict):
        return entry.get('gene', None)
    else:
        return None

df_mapping['ensembl_gene_id'] = df_mapping['ensembl'].apply(extract_ensembl_gene_id)

# Keep relevant columns
df_mapping = df_mapping[['query', 'ensembl_gene_id']]

# Remove entries without mapping
df_mapping = df_mapping.dropna(subset=['ensembl_gene_id'])

# Save the mapping to a CSV file (optional)
df_mapping.to_csv('uniprot_to_ensembl_mapping.csv', index=False)

print("Mapping complete. Mapped UniProt IDs to Ensembl Gene IDs.")
```

```python
df_mapping
```

### **Explanation**

- **`querymany` Function**: Queries multiple UniProt IDs at once.
- **`scopes='uniprot'`**: Specifies that the input identifiers are UniProt IDs.
- **`fields='ensembl.gene'`**: Requests Ensembl Gene IDs in the output.
- **`species='human'`**: Limits the query to human genes.

GTEx uses versioned GENCODE IDs (e.g., `ENSG00000141510.16`).

- **MyGene.info** returns Ensembl Gene IDs without version numbers.
- **Adding Version Numbers**: You may need to add version numbers to match GTEx IDs.

```python
# !wget https://ftp.ebi.ac.uk/pub/databases/gencode/Gencode_human/release_26/gencode.v26.annotation.gtf.gz
```

```python
# Load GENCODE annotation
gencode_gtf = 'gencode.v26.annotation.gtf.gz'
gtf_df = pd.read_csv(gencode_gtf, sep='\t', comment='#', header=None, compression='gzip', low_memory=False)

# Assign column names for clarity
gtf_df.columns = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute']

# Filter for 'gene' entries and create a copy to avoid the warning
gtf_genes = gtf_df[gtf_df['feature'] == 'gene'].copy()

def parse_attributes(attr_str):
    attrs = {}
    for attr in attr_str.strip().split('; '):
        if attr:
            key_value = attr.strip().split(' ', 1)
            if len(key_value) == 2:
                key, value = key_value
                attrs[key] = value.strip('"')
    return attrs

# Parse attributes
gtf_genes['attributes_dict'] = gtf_genes['attribute'].apply(parse_attributes)

# Extract versioned and unversioned Ensembl Gene IDs
gtf_genes['ensembl_gene_id_version'] = gtf_genes['attributes_dict'].apply(lambda x: x.get('gene_id'))
gtf_genes['ensembl_gene_id'] = gtf_genes['ensembl_gene_id_version'].apply(lambda x: x.split('.')[0])

print("Extracted Ensembl Gene IDs with versions from GENCODE GTF.")
```

```python
gtf_genes
```

```python
# Keep only relevant columns
gencode_mapping = gtf_genes[['ensembl_gene_id', 'ensembl_gene_id_version']].drop_duplicates()

print("Created mapping from Ensembl Gene IDs to versioned GENCODE IDs.")
```

```python
# Merge your mapping with GENCODE mapping
df_final_mapping = df_mapping.merge(gencode_mapping, on='ensembl_gene_id', how='left')

# Remove entries without versioned IDs
df_final_mapping = df_final_mapping.dropna(subset=['ensembl_gene_id_version'])

print("Merged mappings to get versioned GENCODE IDs.")
df_final_mapping = df_final_mapping[~df_final_mapping.ensembl_gene_id_version.duplicated()]
df_final_mapping
```

## Option A. Query tissue-specific NeDRex PPIs

https://github.com/REPO4EU/nedrex_api_guide

```python
import requests
from tqdm import tqdm

class NeDRexService:
    API_LINK = "https://exbio.wzw.tum.de/repo4eu_nedrex_open/"

def get_all_edges_reviewed_proteins(iid_evidence=["exp"]):
    all_edges = []
    upper_limit = 10000
    offset = 0
    body = {
        "reviewed_proteins": [True],
        "limit": upper_limit,
        "iid_evidence": iid_evidence
    }

    pbar = tqdm(desc="Fetching edges", unit="batch")
    while True:
        body["skip"] = offset
        try:
            response = requests.post(
                url=f"{NeDRexService.API_LINK}/ppi",
                json=body,
                headers={"content-type": "application/json"}
            )
            response.raise_for_status()
            data = response.json()
            all_edges.extend(data)
            
            pbar.update(1)
            pbar.set_postfix_str(f"Total edges so far: {len(all_edges)}")

            if len(data) < upper_limit:
                # No more data
                break
            offset += upper_limit
        except requests.exceptions.RequestException as e:
            print(f"HTTP request failed: {e}")
            break
    pbar.close()
    return all_edges

# Retrieve all reviewed PPIs and extract tissues
edges = get_all_edges_reviewed_proteins(["exp"])
print(f"Number of edges retrieved: {len(edges)}")

all_tissues = set()
for edge in edges:
    # Each edge may have a "tissues" attribute containing a list of tissues.
    for tissue in edge.get("tissues", []):
        all_tissues.add(tissue)

all_tissues_list = sorted(all_tissues)
print(f"Found {len(all_tissues_list)} unique tissues.")

# Print the full list of tissues
for tissue in all_tissues_list:
    print(tissue)
```

```python
import os
import pickle
import requests
from tqdm import tqdm

class NeDRexService:
    API_LINK = "https://exbio.wzw.tum.de/repo4eu_nedrex_open/"

def get_all_edges_reviewed_proteins(iid_evidence=["exp"], cache_file="edges_cache.pkl"):
    # If cache file exists, load from it
    if os.path.exists(cache_file):
        print("Loading edges from cache...")
        with open(cache_file, "rb") as f:
            edges = pickle.load(f)
        print(f"Loaded {len(edges)} edges from {cache_file}")
        return edges
    
    # Otherwise, fetch from API
    all_edges = []
    upper_limit = 10000
    offset = 0
    body = {
        "reviewed_proteins": [True],
        "limit": upper_limit,
        "iid_evidence": iid_evidence
    }

    pbar = tqdm(desc="Fetching edges", unit="batch")
    while True:
        body["skip"] = offset
        try:
            response = requests.post(
                url=f"{NeDRexService.API_LINK}/ppi",
                json=body,
                headers={"content-type": "application/json"}
            )
            response.raise_for_status()
            data = response.json()
            all_edges.extend(data)
            
            pbar.update(1)
            pbar.set_postfix_str(f"Total edges so far: {len(all_edges)}")

            if len(data) < upper_limit:
                # No more data
                break
            offset += upper_limit
        except requests.exceptions.RequestException as e:
            print(f"HTTP request failed: {e}")
            break
    pbar.close()
    
    # Save to cache
    with open(cache_file, "wb") as f:
        pickle.dump(all_edges, f)
    print(f"Saved {len(all_edges)} edges to {cache_file}")

    return all_edges

# Usage example:
edges = get_all_edges_reviewed_proteins(["exp"])
print(f"Total edges retrieved: {len(edges)}")
```

```python
get_all_edges_reviewed_proteins(["exp"])

tissues_of_interest = [
    'Amygdala',
    'Brain',
    'Dorsal root ganglia',
    'Hypothalamus',
    'Salivary gland'
]

# Remove 'uniprot.' prefix for convenience
for e in edges:
    e["memberOne"] = e["memberOne"].replace("uniprot.", "")
    e["memberTwo"] = e["memberTwo"].replace("uniprot.", "")

# Create a DataFrame for easier handling
df = pd.DataFrame(edges)

# Convert lists to comma-separated strings if needed
# Assuming df['tissues'] is a list of tissues:
df['tissues'] = df['tissues'].apply(lambda x: x if isinstance(x, list) else [])
# Now filter by tissues of interest and build sets

tissue_proteins = {}

for tissue in tissues_of_interest:
    # Filter edges that mention this tissue
    tissue_df = df[df['tissues'].apply(lambda ts: tissue in ts)]
    if tissue_df.empty:
        print(f"No edges for tissue: {tissue}")
        tissue_proteins[tissue] = set()
        continue
    
    # Extract the proteins
    proteins = set(tissue_df['memberOne']) | set(tissue_df['memberTwo'])
    tissue_proteins[tissue] = proteins
    print(f"Tissue: {tissue}, Number of proteins: {len(proteins)}")
```

```python
uniprot_refppi = pd.read_csv(os.path.join(nextflow_dir, "tests/uniprot_ppi.csv"), header=None, names=["memberOne", "memberTwo"])
uniprot_refppi
```

```python
tissues_of_interest = [
    'Amygdala',
    'Brain',
    'Dorsal root ganglia',
    'Hypothalamus',
    'Salivary gland'
]

output_dir = "NeDRex/reference_filtered_ppi_by_tissue_protein_set"
os.makedirs(output_dir, exist_ok=True)

for tissue in tissues_of_interest:
    if tissue not in tissue_proteins or not tissue_proteins[tissue]:
        print(f"No protein set available for tissue '{tissue}', skipping.")
        continue
    
    # Get the set of proteins for this tissue
    t_proteins = tissue_proteins[tissue]
    
    # Filter the smaller PPI network so that both memberOne and memberTwo are in t_proteins
    filtered_df = uniprot_refppi[
        uniprot_refppi['memberOne'].isin(t_proteins) &
        uniprot_refppi['memberTwo'].isin(t_proteins)
    ].copy()
    
    if filtered_df.empty:
        print(f"No edges found for tissue '{tissue}' in the smaller network.")
        continue

    # Save without headers and index, just like before
    filename = tissue.replace(" ", "_").replace("(", "").replace(")", "").replace("-", "_").lower() + ".csv"
    output_path = os.path.join(output_dir, filename)
    filtered_df[['memberOne', 'memberTwo']].to_csv(output_path, index=False, header=False)
    
    # Print stats
    unique_proteins = set(filtered_df['memberOne']) | set(filtered_df['memberTwo'])
    print(f"Tissue: {tissue}")
    print(f"Number of proteins in tissue: {len(t_proteins)}")
    print(f"Number of proteins in the filtered subnetwork: {len(unique_proteins)}")
    print(f"Number of edges in the filtered subnetwork: {len(filtered_df)}")
    print(f"Saved to {output_path}\n")
```

### Filtering by tissue specificity

```python
import tspex
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Set the gene IDs as the index
genes_expr_df.set_index('GENCODE_ID', inplace=True)

# Tissues of interest
tissues = [
    'Brain - Cortex',
    'Brain - Hippocampus',
    'Nerve - Tibial',
    'Skin - Not Sun Exposed (Suprapubic)',
    'Skin - Sun Exposed (Lower leg)',
    'Minor Salivary Gland',
    'Esophagus - Mucosa',
    'Whole Blood',
    'Cells - EBV-transformed lymphocytes'
]

# Prepare expression data
expression_data = genes_expr_df[tissues]
expression_data = expression_data.loc[(expression_data > 0).any(axis=1)]
```

### Guidelines for Setting a Threshold

https://apcamargo.github.io/tspex/metrics/

1. **Biological Plausibility**:
    - A lower threshold (e.g., `SPM >= 0.5`) identifies genes that are moderately specific to a tissue but may still be expressed in others.
    - A higher threshold (e.g., `SPM >= 0.8`) identifies genes with high specificity, often limiting results to those expressed almost exclusively in the tissue of interest.
2. **Benchmarking**:
    - Compare results across different thresholds (e.g., `0.5`, `0.6`, `0.8`) to evaluate the number of genes selected and their biological relevance.

```python

# Compute tissue specificity using SPM
tso = tspex.TissueSpecificity(expression_data, method='spm', log=True)
tissue_specificity = tso.tissue_specificity

# Identify tissue-specific genes
threshold = 0.5
tissue_specific_genes = {}

for tissue in tissues:
    specific_genes = tissue_specificity.index[tissue_specificity[tissue] >= threshold].tolist()
    tissue_specific_genes[tissue] = specific_genes
    print(f"{tissue}: {len(specific_genes)} tissue-specific genes")

# Visualize specificity distribution
tissue_spec_melted = tissue_specificity.reset_index().melt(id_vars='GENCODE_ID', var_name='Tissue', value_name='SPM')
plt.figure(figsize=(12, 6))
sns.violinplot(x='Tissue', y='SPM', data=tissue_spec_melted, inner=None, scale='count', color='C0')
plt.xticks(rotation=90)
plt.title('Distribution of SPM Tissue Specificity Scores')
plt.ylabel('SPM Score')
plt.xlabel('Tissue')
plt.tight_layout()
plt.show()
```

Filtering the PPI by tissue_specific_genes

```python
import os

for tissue in tissues:
    TS_proteins = set(df_final_mapping[df_final_mapping.ensembl_gene_id_version.isin(set(tissue_specific_genes[tissue]))]['query'])
    # Filter PPI where both source and target are in the tissue-specific proteins set
    filtered_ppi = og_ppi[
        og_ppi['source'].isin(TS_proteins) & 
        og_ppi['target'].isin(TS_proteins)
    ]
    
    # Print some stats
    print(f"\nOriginal PPI edges: {len(og_ppi)}")
    print(f"Filtered PPI edges ({tissue}): {len(filtered_ppi)}")
    print(f"Number of tissue-specific ({tissue}) proteins: {len(TS_proteins)}")

    tissue_specific_PPI_dir = f"GTEx/tissue_specific_PPI_{threshold}"
    if not os.path.exists(tissue_specific_PPI_dir):
        os.makedirs(tissue_specific_PPI_dir, exist_ok=True)
         
    filtered_ppi.to_csv(f"{tissue_specific_PPI_dir}/PPI_{tissue.replace(' ', '_').replace('(', '').replace(')', '')}.csv", header=False, index=False)
```
