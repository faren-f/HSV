```python
import pandas as pd
from pathlib import Path
from typing import List
```

```python
base_dir = Path("/home/bbc8731/HSV1-2_pipeline/3_module_expansion/results_full_PPI")
module_dirs = [d for d in base_dir.glob("*") if d.is_dir() and (d.name.startswith('BP_') or d.name.startswith('CC_'))]
methods = ['diamond', 'domino', 'robust', 'robust_bias_aware', 'rwr', 'no_tool']
tissues = ['brain', 'hypothalamus', 'amygdala', 'dorsal_root_ganglia', 'salivary_gland', 'uniprot_ppi']
```

```python
for i in module_dirs:
    nodes_dir = i / 'modules' / 'tsv_nodes'
    Category = i.name
    for tissue in tissues: 
        for method in methods:
            node_file = nodes_dir / f"{Category.split('_', 1)[1]}.{tissue}.{method}.nodes.tsv"
            proteins = pd.read_csv(node_file, sep='\t')            
```
