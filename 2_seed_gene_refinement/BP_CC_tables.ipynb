```python
import matplotlib.pyplot as plt

# -------- Data --------
col_labels_1 = ["Biological process", "HSV-1 proteins", "Unique human partners"]
table1_data = [
    ["Replication and transcription", 19, 162],
    ["Assembly and packaging", 11, 44],
    ["Entry and uncoating", 6, 63],
    ["Egress and envelopment", 19, 205],
    ["Immune evasion", 6, 168],
    ["Uncharacterized", 6, 23],
    ["Unique HSV-1", 67, ""]
]

col_labels_2 = ["Cell compartment", "HSV-1 proteins", "human protein partners"]
table2_data = [
    ["Host Nucleus", 28, 312],
    ["Host Membrane", 16, 197],
    ["Host Cytoplasm", 4, 42],
    ["Virion Components", 9, 52],
    ["Unique HSV-1", 57, ""]
]

# -------- Compact Side-by-Side Figure --------
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # small width + height
fig.patch.set_facecolor("white")

def make_table(ax, data, columns, title, header_color):
    ax.axis("off")
    
    table = ax.table(
        cellText=data,
        colLabels=columns,
        loc="center",
        cellLoc="center"
    )
    
    table.auto_set_font_size(False)
    table.set_fontsize(7)     # very small font
    table.scale(1.0, 1.1)     # tight scaling

    # Header styling
    for col in range(len(columns)):
        table[0, col].set_facecolor(header_color)
        table[0, col].set_text_props(weight='bold', color='white')

    # Zebra rows
    for row in range(1, len(data) + 1):
        color = "#f5f5f5" if row % 2 == 0 else "white"
        for col in range(len(columns)):
            table[row, col].set_facecolor(color)

    # Left-align first column
    for row in range(1, len(data) + 1):
        table[row, 0]._loc = 'left'

    # ax.set_title(title, fontsize=9, weight='bold', pad=6)

make_table(
    axes[0],
    table1_data,
    col_labels_1,
    "Biological Process",
    header_color="#2c7fb8"
)

make_table(
    axes[1],
    table2_data,
    col_labels_2,
    "Cell Compartment",
    header_color="#41ab5d"
)

plt.tight_layout(w_pad=1)

plt.savefig(os.path.join(plot_dir, "seed_genes_tables.pdf"), format="pdf", bbox_inches="tight")
plt.savefig(os.path.join(plot_dir, "seed_genes_tables.png"), format="png", dpi=300, bbox_inches="tight")
plt.close()

print("Saved as seed_genes_tables.pdf")
```
