```python
import os
import glob
import pandas as pd
import numpy as np
from pathlib import Path, PosixPath
import matplotlib.pyplot as plt
```

```python
workdir = os.path.dirname(os.getcwd())
input_dir = Path(f"{workdir}/5_module_evaluation/data/")
paths = glob.glob(os.path.join(input_dir, "*_module_enrichment.csv"))
plot_dir = "/home/bbc8731/HSV/5_module_evaluation/data/plot"
```

```python
adj_pval = []
module_size = []
fold_enrichment = []
for path in paths:
    df = pd.read_csv(path)

    # set module as index
    df = df.set_index("module")

    # extract only category name (before first "/")
    df.index = df.index.str.split("/").str[0]

    adj_pval.append(df["fdr"])
    module_size.append(df["module_size"])
    fold_enrichment.append(df["fold_enrichment"])

adj_pval = pd.concat(adj_pval, axis=1)
module_size = pd.concat(module_size, axis=1)
fold_enrichment = pd.concat(fold_enrichment, axis=1)


# name columns properly
adj_pval.columns = [f"thr_{i+1}" for i in range(adj_pval.shape[1])]
module_size.columns = [f"thr_{i+1}" for i in range(module_size.shape[1])]
fold_enrichment.columns = [f"thr_{i+1}" for i in range(fold_enrichment.shape[1])]
```

```python
# extract column
adj_pval_thr4 = adj_pval["thr_4"]
module_size_thr4 = module_size["thr_4"]
fold_enrichment_thr4 = fold_enrichment["thr_4"]

# sort by p-value
sorted_idx = adj_pval_thr4.sort_values().index

adj_pval_thr4 = adj_pval_thr4.loc[sorted_idx]
module_size_thr4 = module_size_thr4.loc[sorted_idx]
fold_enrichment_thr4 = fold_enrichment_thr4.loc[sorted_idx]

# transform
data = -np.log10(adj_pval_thr4)
sig_threshold = -np.log10(0.05)

fig, ax = plt.subplots(figsize=(8, 6))

colors = [
    "steelblue" if v >= sig_threshold else "lightgray"
    for v in data.values
]

bars = ax.barh(data.index, data.values, color=colors)

# add module size + fold enrichment
for bar, size, fe in zip(bars, module_size_thr4, fold_enrichment_thr4):
    ax.text(
        bar.get_width() + 0.1,
        bar.get_y() + bar.get_height() / 2,
        f"n={size}, FE={fe:.2f}",
        va="center"
    )

ax.axvline(sig_threshold, linestyle="--", color="red", linewidth=1.5)

ax.set_xlabel("-log10(FDR)")
ax.set_title("Module Enrichment (thr_4)")
plt.tight_layout()
# plt.show()

plt.savefig(os.path.join(plot_dir, "Module Enrichment_thr4.pdf"), format="pdf")
plt.close()
```

```python
adj_pval = adj_pval[["thr_4", "thr_5", "thr_6"]]

data = -np.log10(adj_pval)
sig_threshold = -np.log10(0.05)

fig, ax = plt.subplots()
im = ax.imshow(data, aspect='auto')

# mark significant cells
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        if data.iloc[i, j] >= sig_threshold:
            ax.add_patch(
                plt.Rectangle(
                    (j - 0.5, i - 0.5),
                    1, 1,
                    fill=False,
                    edgecolor='red',
                    linewidth=1.5
                )
            )

ax.set_xticks(range(len(adj_pval.columns)))
ax.set_xticklabels(adj_pval.columns, rotation=45)
ax.set_yticks(range(len(adj_pval.index)))
ax.set_yticklabels(adj_pval.index)

plt.colorbar(im, label='-log10(FDR)')
plt.xlabel("Threshold")
plt.ylabel("Module")
plt.tight_layout()
plt.show()
```
